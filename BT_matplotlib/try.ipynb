```python
# Đọc file
df = pd.read_csv("Mall_Customers_2.csv")
```

```python
# Thay missing data bằng giá trị trung bình
df['Age'].fillna(df['Age'].mean(), inplace = True)
df['Spending Score (1-100)'].fillna(df['Spending Score (1-100)'].mean(), inplace = True)
```

```python
# Lấy ra list tuổi và điểm
age = df['Age']
score = df['Spending Score (1-100)']
```

```python
# Lấy màu
color = ["blue" if i == "Male" else "pink" for i in df["Genre"]]
```

```python
# Vẽ đồ thị
for i in range(len(color)):
  plt.scatter(age[i], score[i], color = color[i])
plt.title('Scatter\n')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)');
```

```python
salary = df['Annual Income (k$)']
```

```python
plt.plot(age, salary, c='green')
plt.title('Line\n')
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")

# Lấy mũi tên chir vào max
id = salary.argmax()
plt.annotate('Max', xy=(age[id], salary[id]), xytext=(age[id]+10, salary[id]+10), arrowprops=dict(facecolor='red'));
```

```python
df1 = pd.read_csv('Social_Network_Ads_2.csv')
```

```python
df1['Age'].fillna(df1['Age'].mean(), inplace = True)
df1['EstimatedSalary'].fillna(df1['EstimatedSalary'].mean(), inplace = True)
df1['Purchased'].fillna(df1['Purchased'].mean(), inplace = True)
```

```python
# Tạo list status
status = ['Chua mua' if (i == 0) else 'Da mua' for i in df1['Purchased']]
# Đồi tên cột Purchased -> Status
df1.rename(columns = {'Purchased' : 'Status'})
# Lấy giá trị cho cột status
df1['Status'] =  status
```

```python
age = df1['Age']
sta = df1['Status']
est = df1['EstimatedSalary']
```

```python
color = ['red' if s == 'Chua mua' else 'green' for s in df1['Status']]
```

```python
for i in range(len(color)) :
    plt.bar(age[i],est[i], color = color[i])

plt.title('Bar\n')
plt.xlabel("Age")
plt.ylabel("EstimatedSalary");
```

```python
color1 = ["violet" if i == "Da mua" else "Red" for i in df1["Status"]]
```

```python
for i in range(len(color)):
    plt.scatter(age[i],est[i], c = color1[i])

plt.title('Scatter\n')
plt.xlabel("Age")
plt.ylabel("EstimatedSalary");
```
